```python
import pandas as pd
import requests
import random
```

```python
df=pd.read_csv("./CleanedData/CleanedHousingRecommenderDataset.csv")
attributes=['period_end','county','state','state_code','median_ppsf','tax_burden','medianage','cost_of_living_rank']
df=df[attributes]
df['tax_burden'] = df['tax_burden'].astype(str).apply(lambda x: x.split("%")[0]).astype(float)
df
```

```python
df.columns
```

```python
forcasted_data=pd.read_csv("./CleanedData/ppsf_forecast.csv")
#Housing Recommender System\DataExtraction\CleanedData\ppsf_forecast.csv
df_subset=df[['county','state_code','tax_burden', 'medianage', 'cost_of_living_rank',]]
df_unique = df_subset.groupby(by=['county','state_code'], as_index=False).first()
forcasted_data=forcasted_data.drop(['month','year'],axis=1)
forcasted_data.rename(columns={'date':'period_end','state':'state_code'}, inplace=True)
df_predicted = forcasted_data.merge(df_unique, on=["state_code","county"], how = 'left')
df_predicted
#forcasted_data
```

```python
def getSchoolRatings(df):
    elem_school=[random.randint(1,10) for i in range(0,len(df))]
    middle_school=[random.randint(1,10) for i in range(0,len(df))]
    high_school=[random.randint(1,10) for i in range(0,len(df))]
    df['elementary_school_rating']=elem_school
    df['middle_school_rating']=middle_school
    df['high_school_rating']=high_school
    return df


df_subset=df[['county','state_code']]
df_unique = df_subset.groupby(by=['county','state_code'], as_index=False).first()
school_ratings=getSchoolRatings(df_unique)
df_final = df_final.merge(school_ratings, on=["state_code","county"], how = 'left')
#dfinal=dfinal[df.columns]
df_final
```
